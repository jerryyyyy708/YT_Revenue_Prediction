```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
from sklearn import preprocessing

#dataset owner:jerryyyyy708
#https://youtube.com/channel/UCLi9jBe_Jzo1vPjULRpCj7Q
data="dataset/YT_revenue.csv"

#load dataset and discard useless features
df=pd.read_csv(data,sep=',')
cols = [col for col in df if col != 'YouTube ad revenue (TWD)']
cols.append('YouTube ad revenue (TWD)')
df = df[cols]
df=df.drop(columns=['Video title'])
df=df.drop(columns=['Video publish time'])
df=df.drop(columns=['Video'])
#df=df.drop(columns=['Dislikes'])
df=df.drop(columns=['Comments added'])
df=df.drop(columns=['Shares'])
df=df.drop(columns=['Subscribers gained'])
#df=df.drop(columns=['Likes'])
#df=df.dropna()
#df=df.drop(df.index[-1])
#df=df.drop(df.index[0])
#df.drop(df[df['YouTube ad revenue (TWD)']==0].index,inplace=True)
df=shuffle(df,random_state=1)
labels=list(df['YouTube ad revenue (TWD)'])
df = df.drop(columns= ['YouTube ad revenue (TWD)'])
df
```

```python
df1 = pd.read_csv('dataset/YT_revenue.csv', encoding="UTF-8")
df2 = pd.read_csv('dataset/append.csv', encoding = "UTF-8")
line1 = set(df1['Video'])
line2 = set(df2['Video'])
videos = list(line1 & line2)
df2 = df2[df2['Video'].isin(videos)]
#keep the videos that are in both set for df2

column1 = [i for i in df1]
column2 = [i for i in df2]
drop = list(set(column1) & set(column2) - set(['Video']))
#drop = drop - ['Video']
df2 = df2.drop(columns=drop)
df = pd.merge(df1, df2, on = 'Video', how = 'inner')
df['Average view duration'] = df['Average view duration'].str.split(':').apply(lambda x: int(x[0]) * 3600 + int(x[1]) * 60 + int(x[2]))
df.to_csv('dataset/YT_Revenue_v2.csv', index = False)

df
```

```python
#Linear Regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

features=df.iloc[:,:]
target=labels

#normalize
features=preprocessing.scale(features)

X_train,X_test, y_train, y_test =train_test_split(features,target,test_size=0.3,random_state=1)
model=RandomForestRegressor(n_estimators = 20)
model.fit(X_train, y_train)
y_predict=model.predict(X_test)

average_err=0
average_rev=0
average_predict=0
revenue=[]

#calculate error rate
for i in range(len(y_predict)):
    revenue.append(y_test[i])
    error=y_predict[i]-y_test[i]
    average_predict+=y_predict[i]
    if error < 0:
        error=(-error)
    average_err+=error
    average_rev+=y_test[i]
average_err=average_err/len(y_predict)
average_rev=average_rev/len(y_predict)
average_predict=average_predict/len(y_predict)
#show result
print("average predict $:",average_predict)
print("average revenue $:",average_rev)
print("average error $:",average_err)
print("average error %:",average_err/average_rev)
```
